# OneHot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ohe = OneHotEncoder(drop='first', handle_unknown='ignore')
onehot_columns = ['BusinessTravel', 'Department', 'Gender', 'MaritalStatus', 'OverTime']

# column transformer 
preprocessing = ColumnTransformer(
    transformers=[
        ('OneHotEncoding', ohe, onehot_columns)
    ],
    remainder='passthrough'
)

# fit and transform
transformed_array = preprocessing.fit_transform(df)

# Get correct encoded column names
encoded_cols = preprocessing.named_transformers_['OneHotEncoding'].get_feature_names_out(onehot_columns)

# Get non encoded column names
non_encode_cols = [feature for feature in df.columns if feature not in onehot_columns]

# Total columns
total_columns = list(encoded_cols) + non_encode_cols
```

```python
new_df = pd.DataFrame(transformed_array, columns=total_columns)
```

```python
new_df.select_dtypes(include=['object']).columns
```

```python
new_df['Attrition'].head()
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

label_encode_features = ['EducationField', 'JobRole', 'Attrition']

for feature in label_encode_features:
    le = LabelEncoder()
    new_df[feature] = le.fit_transform(new_df[feature])
```

```python
new_df['Attrition'].head()
```

# Make all the features into numeric

```python
new_df = new_df.apply(pd.to_numeric, errors='ignore')
```

# Model training

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

X = new_df.drop('Attrition', axis=1)
y = new_df['Attrition']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.30)

smote = SMOTE(random_state=42)
X_train, y_train = smote.fit_resample(X_train, y_train)
```

# Fix the outliers

```python
def fix_outlier(feature, dataset):
    Q1 = dataset[feature].quantile(0.25)
    Q3 = dataset[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    dataset[feature] = np.clip(dataset[feature], lower, upper)

for feature in numerical_features:
    fix_outlier(feature, X_train)
    fix_outlier(feature, X_test)
```

```python
X_train[numerical_features].skew()
```

# Fix the skweness of numerical features

```python
from sklearn.preprocessing import PowerTransformer, StandardScaler

# Identify skewed numerical features
skewness = X_train[numerical_features].skew()
skewed_features = skewness[(skewness > 1) | (skewness < -1)].index.tolist()
less_skewed_features = skewness[(skewness <= 1) & (skewness >= -1)].index.tolist()

# Apply PowerTransformer only on skewed features
pt = PowerTransformer(method='yeo-johnson')
X_train[skewed_features] = pt.fit_transform(X_train[skewed_features])
X_test[skewed_features] = pt.transform(X_test[skewed_features])
```

# Apply standard scaler on less skewd features

```python
# Apply StandardScaler on the rest of the numerical features
scaler = StandardScaler()
X_train[less_skewed_features] = scaler.fit_transform(X_train[less_skewed_features])
X_test[less_skewed_features] = scaler.transform(X_test[less_skewed_features])
```

# Feature selection

```python
def select_features(dataset, threshold):
    corr_set = set()
    corr_matrix = dataset.corr()

    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if np.abs(corr_matrix.iloc[i, j]) > threshold:
                col_name = corr_matrix.columns[i]
                corr_set.add(col_name)
    return corr_set

del_cols = select_features(X_train, 0.8)
X_train.drop(del_cols, axis=1, inplace=True)
X_test.drop(del_cols, axis=1, inplace=True)
```

# Save the dataset

```python
X_train.to_csv('X_train.csv', index=False)
X_test.to_csv('X_test.csv', index=False)
y_train.to_csv('y_train.csv', index=False)
y_test.to_csv('y_test.csv', index=False)
```
