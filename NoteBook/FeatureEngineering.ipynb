# OneHot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ohe = OneHotEncoder(drop='first', handle_unknown='ignore')
onehot_columns = ['BusinessTravel', 'Department', 'Gender', 'MaritalStatus', 'OverTime']

# column transformer 
preprocessing = ColumnTransformer(
    transformers=[
        ('OneHotEncoding', ohe, onehot_columns)
    ],
    remainder='passthrough'
)

# fit and transform
transformed_array = preprocessing.fit_transform(df)

# Get correct encoded column names
encoded_cols = preprocessing.named_transformers_['OneHotEncoding'].get_feature_names_out(onehot_columns)

# Get non encoded column names
non_encode_cols = [feature for feature in df.columns if feature not in onehot_columns]

# Total columns
total_columns = list(encoded_cols) + non_encode_cols
```

```python
new_df = pd.DataFrame(transformed_array, columns=total_columns)
```

# Make all the features into numeric

```python
new_df = new_df.apply(pd.to_numeric, errors='ignore')
```

# Model training

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

X = new_df.drop('Attrition', axis=1)
y = new_df['Attrition']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.30)

smote = SMOTE()
X_train, y_train = smote.fit_resample(X_train, y_train)
```

# Fix the outliers

```python
def fix_outlier_train(feature):
    iqr = X_train[feature].quantile(0.75) - X_train[feature].quantile(0.25)
    upper_border = X_train[feature].quantile(0.75) + (iqr*1.5)
    lower_border = X_train[feature].quantile(0.25) - (iqr*1.5)
    X_train[feature] = np.clip(X_train[feature], lower_border, upper_border)
    return upper_border, lower_border

outlier_thresholds = {feature: fix_outlier_train(feature) for feature in numerical_features}


def fix_outlier_test(feature, threshold):
    upper_border, lower_border = threshold[feature]
    X_test[feature] = np.clip(X_test[feature], lower_border, upper_border)

for feature in numerical_features:
    fix_outlier_test(feature, outlier_thresholds)
```

```python
X_train[numerical_features].skew()
```

# Fix the skweness of numerical features

```python
from scipy.stats import boxcox

train_skew_features = X_train[numerical_features].skew()
test_skew_features = X_test[numerical_features].skew()
transforms_features = []
boxcox_lambdas = {}

# Train data
for feature in numerical_features:
    skewness = train_skew_features[feature]

    if skewness>1:
        if (X_train[feature]>0).all():
            X_train[feature] = np.log1p(X_train[feature])
        else:
            X_train[feature] = np.sqrt(X_train[feature] - X_train[feature].min() + 1)
        transforms_features.append(feature)
    elif skewness<-1:
        X_train[feature], boxcox_lambda = boxcox(X_train[feature] - X_train[feature].min() + 1)
        boxcox_lambdas[feature] = boxcox_lambda
        transforms_features.append(feature)

# Test data
for feature in numerical_features:
    skewness = test_skew_features[feature]

    if skewness>1:
        if (X_test[feature]>0).all():
            X_test[feature] = np.log1p(X_test[feature])
        else:
            X_test[feature] = np.sqrt(X_test[feature] - X_test[feature].min() + 1)
    elif skewness<-1:
        X_test[feature], boxcox_lambda = boxcox(X_test[feature] - X_test[feature].min() + 1, boxcox_lambdas[feature])
```

# Apply standard scaler on less skewd features

```python
less_skewd_features = [feature for feature in numerical_features if feature not in transforms_features]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train[less_skewd_features] = scaler.fit_transform(X_train[less_skewd_features])
X_test[less_skewd_features] = scaler.transform(X_test[less_skewd_features])
```

# Feature selection

```python
def select_features(dataset, threshold):
    corr_set = set()
    corr_matrix = dataset.corr()

    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if np.abs(corr_matrix.iloc[i, j]) > threshold:
                col_name = corr_matrix.columns[i]
                corr_set.add(col_name)
    return corr_set

del_cols = select_features(X_train, 0.6)
X_train.drop(del_cols, axis=1, inplace=True)
X_test.drop(del_cols, axis=1, inplace=True)
```
